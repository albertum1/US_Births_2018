```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('abalone.csv')
```

```python
df['waste'] = df['Whole weight'] - df['Shucked weight'] - df['Viscera weight'] - df['Shell weight']
```

```python
df['waste'] = df['waste']/df['Whole weight']
```

```python
df
```

```python
sns.distplot(df['waste'])
```

```python
df['rings_log'] = np.log(df['Rings'])
```

```python
sns.scatterplot(x='rings_log', y='Whole weight', data= df)
```

```python
df['whole_weight_log'] = np.log(df['Whole weight'])
```

```python
sns.scatterplot(x='rings_log', y='whole_weight_log', data= df)
```

```python
sns.scatterplot(x='Rings', y='whole_weight_log', data= df)
```

```python
df['sqrt_weight'] = np.sqrt(df['Whole weight'])
```

```python
sns.scatterplot(x='Rings', y='sqrt_weight', data= df)
```

```python
df['sqrt_rings'] = np.sqrt(df['Rings'])
```

```python
sns.scatterplot(x='sqrt_rings', y='Whole weight', data= df)
```
